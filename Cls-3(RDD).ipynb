```python
##Map In textfile

def fun(lines):
    lines=lines.lower()
    lines=lines.split()
    return lines

split_RDD=z.map(fun)   
```

```python
split_RDD.collect()
```

```python
split_RDD.take(10)
```

```python
##Flat Map

def func(lines):
    lines=lines.lower()
    lines=lines.split()
    return lines
    
split_RDD2=z.flatMap(func)
```

```python
split_RDD2.collect()
```

```python
split_RDD2.take(10)
```

```python
##stopword

stopword=['is','an','to','so','the', 'of', 'its']

```

```python
RDD1=split_RDD2.filter(lambda x: x not in stopword)
RDD1.collect()
```

```python
RDD1.take(10)
```

```python
###Find word start with :"i"

Rdd2=split_RDD2.filter(lambda x: x.startswith('i'))
Rdd2.collect()
```

```python
Rdd2.distinct().collect()
```

```python
#Union/intersection/subtract/cartesian

m=x=spark.sparkContext.parallelize([1,2,3,4,5,6,7,8])
n=x=spark.sparkContext.parallelize([3,4,5,8,10])
```

```python
m.union(n).collect()
```

```python
m.union(n).distinct().collect()  #Unique only
```

```python
m.intersection(n).collect()
```

```python
m.subtract(n).collect()
```

```python
m.cartesian(n).collect()
```
